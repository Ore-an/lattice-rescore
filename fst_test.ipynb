```python
with open('examples/1.lat.sym') as f:
    ln = f.readlines()
```

```python
lat = {}
tmp_lat = []
uttid_found = False
for line in ln:
    line = line.strip()
    if not uttid_found:
        uttid = line
        uttid_found = True
    else:
        if len(line) == 0:
            lat[uttid] = tmp_lat
            uttid_found = False
        else:
            tmp_lat.append(line)
```
